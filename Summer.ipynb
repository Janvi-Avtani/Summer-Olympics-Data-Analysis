# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
fd = pd.read_csv("summer.csv")
```

```python
fd
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(fd['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data= []
for sport in fd['Sport'].unique():
    data.append([sport , len(fd[(fd['Sport']== sport) & (fd["Medal"]=="Gold")])])
    
pd.DataFrame(data, columns= ['Sport', 'Gold Medals']).sort_values(by='Gold Medals', ascending=False).head().plot(x = 'Sport', y = 'Gold Medals',color = '#4d79ff', kind = 'bar', figsize = (5,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]
for i in fd['Sport'].unique():
    data.append([i , len(fd[fd['Sport']== i])])
    
pd.DataFrame(data, columns= ['Sport', 'Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sport', y = 'Medals',color = '#4d79ff', kind = 'bar', figsize = (5,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data=[]
for i in fd['Athlete'].unique():
    data.append([i , len(fd[fd['Athlete']== i])])
    
pd.DataFrame(data, columns= ['Athlete', 'Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Athlete', y = 'Medals',color = '#4d79ff', kind = 'bar', figsize = (5,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data= []
for i in fd['Athlete'].unique():
    data.append([i , len(fd[(fd['Athlete']== i) & (fd["Medal"]=="Gold")])])
    
pd.DataFrame(data, columns= ['Athlete', 'Gold Medals']).sort_values(by='Gold Medals', ascending=False).head().plot(x = 'Athlete', y = 'Gold Medals',color = '#4d79ff', kind = 'bar', figsize = (5,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data= []
for i in fd['Year'].unique():
    if(len(fd[(fd['Year']== i) & (fd["Medal"]=="Gold") &(fd['Country']=="IND")]) >0):
        data.append([i , len(fd[(fd['Year']== i) & (fd["Medal"]=="Gold") &(fd['Country']=="IND")])])
        
pd.DataFrame(data, columns= ['Year', 'Gold Medals']).sort_values(by='Year', ascending=True).head(1)
 
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]
for i in fd['Event'].unique():
    data.append([i , len(fd[fd['Event']== i])])
    
pd.DataFrame(data, columns= ['Event', 'Players']).sort_values(by='Players', ascending=False).head().plot(x = 'Event', y = 'Players',color = '#4d79ff', kind = 'bar', figsize = (5,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data= []
for i in fd['Sport'].unique():
    data.append([i , len(fd[(fd['Sport']== i) & (fd["Medal"]=="Gold") & (fd["Gender"]=="Women")])])
    
pd.DataFrame(data, columns= ['Sport', 'Female Gold Medalists']).sort_values(by='Female Gold Medalists', ascending=False).head().plot(x = 'Sport', y = 'Female Gold Medalists',color = '#4d79ff', kind = 'bar', figsize = (5,5))
```
